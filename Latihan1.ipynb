1. 5 Mahasiswa

```python
mahasiswa = ["Udin", "Andika", "Gita", "Dio", "Putri"]
print("List Mahasiswa:")
print(mahasiswa)
```

```python
2. Tambah 1 Nama Baru dengan append
```

```python
mahasiswa.append("Budi")
print("List yang diperbarui")
print(mahasiswa)
for i in range(len(mahasiswa)):
    print(f"Mahasiswa index ke-{i}: {mahasiswa[i]}")
```

3.Hapus 1 Nama (putri)

```python
print("List Awal")
print(mahasiswa)

mahasiswa.remove("Putri")
print("List yang diperbarui")
print(mahasiswa)
```
